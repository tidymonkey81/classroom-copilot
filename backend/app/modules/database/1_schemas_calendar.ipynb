```python
# Functions to extract data from Excel planner file and handle it

def extract_term_dates(df):
    term_dates = []
    num_terms = int(df[df['Identifier'] == 'AcademicTerms']['Data'].iloc[0])
    for term in range(1, num_terms + 1):
        start_date_identifier = f'Term{term}Start'
        end_date_identifier = f'Term{term}End'
        start_date = df[df['Identifier'] == start_date_identifier]['Data'].iloc[0]
        end_date = df[df['Identifier'] == end_date_identifier]['Data'].iloc[0]
        term_dates.append((start_date, end_date))
    return term_dates

def extract_term_break_dates(df):
    term_break_dates = []
    break_identifiers = [id for id in df['Identifier'].unique() if 'BreakStart' in id]
    for identifier in break_identifiers:
        break_name = identifier.replace('Start', '')
        start_date = df[df['Identifier'] == break_name + 'Start']['Data'].iloc[0]
        end_date = df[df['Identifier'] == break_name + 'End']['Data'].iloc[0]
        term_break_dates.append((start_date, end_date))
    return term_break_dates

def extract_period_times(df):
    # Create a dictionary to hold period times
    period_times = {}
    
    # Iterate over each row in the dataframe to find period start and end times
    for index, row in df.iterrows():
        identifier = row['Identifier']
        # Check if the identifier is for a period start or end time
        if '.PeriodStartTime' in identifier or '.PeriodEndTime' in identifier:
            # Extract the period name and whether it's a start or end time
            parts = identifier.split('.')
            period_name = parts[1]  # e.g., "Reg", "1", "Brk"
            time_type = 'start' if 'StartTime' in identifier else 'end'
            
            # Check the data type of the row's data
            time_data = row['Data']
            if isinstance(time_data, time):
                time_obj = time_data
            else:
                # Convert the time to a datetime.time object, assuming the time is in HH:MM format
                time_obj = datetime.strptime(time_data, '%H:%M').time()
            
            # Initialize a dictionary for the period if it's not already there
            if period_name not in period_times:
                period_times[period_name] = {}
            
            # Assign the start or end time to the period
            period_times[period_name][time_type] = time_obj
            
    return period_times

def extract_sims_timetable_data(df):
    # Create a dictionary to hold the timetable data
    timetable_data = {}
    
    # Iterate over each row in the dataframe to find timetable data
    for index, row in df.iterrows():
        logging.debug(f'Processing row {index}')
        # Create a dictionary to hold the timetable data for this row
        row_data = {}
        # Extract the period identifier
        timetabled_period_identifier = row['Period']
        period_type = str(timetabled_period_identifier)[0:1]
        day = str(timetabled_period_identifier)[1:4]
        period_number = str(timetabled_period_identifier)[5:].replace(':', '')
        period_timetable_code = period_type + day + period_number
        staff_code = row['Staff Code']
        staff_name = row['Teacher']
        room = row['Room']
        subject = row['Subject']
        subject_code = row['Subject code']
        class_code = row['Class']
        class_schweme_name = row['Scheme Name']
        
        period_name = staff_code + '_' + period_type + day + str(period_number)

        # Add the data to the row data dictionary
        row_data['period_type'] = period_type
        row_data['day'] = day
        row_data['period_number'] = period_number
        row_data['period_timetable_code'] = period_timetable_code
        row_data['period_name'] = period_name
        row_data['staff_code'] = staff_code
        row_data['staff_name'] = staff_name
        row_data['room'] = room
        row_data['subject'] = subject
        row_data['subject_code'] = subject_code
        row_data['class_code'] = class_code
        row_data['class_schweme_name'] = class_schweme_name

        # Add the row data to the timetable data dictionary
        timetable_data[period_name] = row_data

    return timetable_data

# Function to populate days array
def populate_days_array(df):
    days = []
    for _, row in df.iterrows():
        day_type = 'School Day' if pd.isna(row['DayModifier']) else \
                   'Holiday' if row['DayModifier'] == 'H' else \
                   'Staff Day' if row['DayModifier'] == 'X' else \
                   'Day Off Timetable' if row['DayModifier'] == 'C' else 'Unknown'
        day_info = {
            'Date': row['Date'],
            'Day': row['Day'],
            'DayType': day_type,
            'DayAutoAgenda': row['DayAutoAgenda'],
            'DayAgendaHeading': row['DayAgendaHeading'],
            'DayAgendaNotes': row['DayAgendaNotes']
        }
        days.append(day_info)
    logging.debug(f"Days array: {days}")
    return days

# Function to populate weeks array
def populate_weeks_array(df):
    weeks = []
    for _, row in df.iterrows():
        week_end_date = row['WeekStartDate'] + timedelta(days=6)  # Assuming a week is 7 days long
        week_type = row['WeekType']
        week_info = {
            'WeekStartDate': row['WeekStartDate'],
            'WeekEndDate': week_end_date,
            'WeekType': week_type,
            'WeekAgendaHeading': row['WeekAgendaHeading'],
            'WeekAgendaNotes': row['WeekAgendaNotes']
        }
        weeks.append(week_info)
    logging.debug(f"Weeks array: {weeks}")
    return weeks

# Function to convert a date string to a date object
def convert_excel_date(excel_date):

    if pd.isna(excel_date) or excel_date == 'Null':
        logging.debug(f"Doing nothing. Excel date is NaN or Null: {excel_date}")
        return None
    if isinstance(excel_date, datetime):
        logging.debug(f"Sending it back. Excel date is already a datetime: {excel_date}")
        return excel_date.date()
    else:
        # Assuming excel_date is a fraction of the day
        logging.debug(f"Making assumption. Excel date is a fraction of the day: {excel_date}")
        return (datetime(1899, 12, 30) + timedelta(days=excel_date)).date()
    
# Function to convert a time string to a time object
def convert_to_date(date_str):
    
    try:
        logging.debug(f"Converting {date_str} to date object...")

        return pd.to_datetime(date_str, dayfirst=True).date()  # Assuming day comes first in your date format
    
    except ValueError:
        logging.error(f"Error converting {date_str} to date object.")
        
        return None
```

```python
# Calendar helper functions

# Function to convert a camel case string to snake case and remove the 'Node' suffix
def camel_to_snake_strip_nodes_end(name):
    # Replace capital letters with underscores and lowercase them
    name = re.sub(r'(?<!^)(?=[A-Z])', '_', name).lower()
    # Remove the 'Node' suffix
    name = name.replace('_node', '')
    return name

# Helper function to check if a date falls within a range
def date_in_range(date, start_date, end_date):
    logging.pedantic(f"Checking if {date} is between {start_date} and {end_date}...")
    logging.pedantic(f"Return will be {start_date <= date <= end_date}")
    return start_date <= date <= end_date

# Helper function to check if a datetime falls within a range
def datetime_in_daterange(check_datetime, start_date, end_date):
    # Convert neo4j.time.Date objects to datetime.date
    start_date = datetime(start_date.year, start_date.month, start_date.day).date()
    end_date = datetime(end_date.year, end_date.month, end_date.day).date()

    # Combine dates with earliest and latest time to create datetime objects
    start_datetime = datetime.combine(start_date, time.min)
    end_datetime = datetime.combine(end_date, time.max)

    logging.pedantic(f"Checking if {check_datetime} is between {start_datetime} and {end_datetime}...")
    logging.pedantic(f"Return will be {start_datetime <= check_datetime <= end_datetime}")
    return start_datetime <= check_datetime <= end_datetime
```

```python
# Create calendar nodes
    
def create_calendar_dicts(start_year, days_data, weeks_data):

    logging.info(f"Creating calendar for {start_year}...")

    start_date = date(start_year, 1, 1)
    end_date = date(start_year + 1, 12, 31)

    week_nodes_dict = {}
    day_nodes_dict = {}

    # Convert days_data and weeks_data into dictionaries for easy lookup
    days_data_dict = {pd.to_datetime(d['Date']).date(): d for d in days_data}
    weeks_data_dict = {pd.to_datetime(w['WeekStartDate']).date(): w for w in weeks_data}

    current_date = start_date

    while current_date <= end_date:

        week_start = current_date - timedelta(days=current_date.weekday())
        week_id = week_start.strftime('%Y-W%W')

        if week_id not in week_nodes_dict:

            logging.debug(f"Creating week node for week {week_id}...")

            week_data = weeks_data_dict.get(week_start) # Get week data if it exists

            week_notes = week_data['WeekAgendaNotes'] if week_data and 'WeekAgendaNotes' in week_data and pd.notna(week_data['WeekAgendaNotes']) else 'General calendar week' # Default to 'General calendar week' if WeekAgendaNotes is not specified
            week_type = week_data['WeekType'] if week_data and 'WeekType' in week_data and pd.notna(week_data['WeekType']) else '' # Default to empty string if WeekType is not specified

            # Create week node
            week_nodes_dict[week_id] = CalendarWeekNode(
                id=week_id,
                start_date=week_start,
                end_date=week_start + timedelta(days=6),
                type=week_type,
                notes=week_notes
            )

        week_type = week_nodes_dict[week_id].week_type

        day_data = days_data_dict.get(current_date)

        if current_date.weekday() in [5, 6]:  # Check if it's a weekend
            day_modifier = 'Weekend'
        else:
            # Check if it's a school day or holiday
            day_modifier = day_data['DayType'] if day_data and 'DayType' in day_data and pd.notna(day_data['DayType']) else 'Calendar Day' # Default to Calendar Day if DayType is not specified     
        
        day_auto_agenda = day_data['DayAutoAgenda'] if day_data and 'DayAutoAgenda' in day_data and pd.notna(day_data['DayAutoAgenda']) else '' # Default to empty string if DayAutoAgenda is not specified
        day_agenda_heading = day_data['DayAgendaHeading'] if day_data and 'DayAgendaHeading' in day_data and pd.notna(day_data['DayAgendaHeading']) else '' # Default to empty string if DayAgendaHeading is not specified
        day_agenda_notes = day_data['DayAgendaNotes'] if day_data and 'DayAgendaNotes' in day_data and pd.notna(day_data['DayAgendaNotes']) else '' # Default to empty string if DayAgendaNotes is not specified

        agenda_notes = day_auto_agenda + '\n' + day_agenda_heading + '\n' + day_agenda_notes
        
        # Create day node
        day_nodes_dict[current_date.strftime('%Y-%m-%d')] = CalendarDayNode(
            date=current_date,
            day=current_date.strftime('%A'),
            type=week_type,
            modifier=day_modifier,
            notes=agenda_notes
        )

        current_date += timedelta(days=1)

    return week_nodes_dict, day_nodes_dict
```

```python
# Create calendar relationships

def create_calendar_relationships(neo4j_driver, node_types_options, has_relationships, part_of_relationships, write=False):
    """
    Create relationships between nodes based on provided options.

    Args:
        neo4j_driver: The Neo4j driver for database connection.
        has_relationships (bool): Whether to create 'HAS' relationships.
        part_of__relationships (bool): Whether to create 'PART_OF' relationships.
        node_types_options (dict): A dictionary specifying which node types to apply the relationships to.
        write (bool): Whether to write changes to the database.
    """
    logging.info(f"Creating calendar relationships...")
    # Fetch nodes from the database based on the node types specified in the options
    if node_types_options['CalendarYearNode']:
        logging.debug(f"Fetching calendar year nodes...")
        calendar_year_nodes = neo4j.find_nodes_by_label(neo4j_driver, "CalendarYear")
        if not calendar_year_nodes:
            logging.error("No calendar year nodes found in the database.")
            return
    if node_types_options['CalendarMonthNode']:
        logging.debug(f"Fetching month nodes...")
        month_nodes = neo4j.find_nodes_by_label(neo4j_driver, "Month")
        if not month_nodes:
            logging.error("No month nodes found in the database.")
            return
    if node_types_options['CalendarWeekNode']:
        logging.debug(f"Fetching week nodes...")
        week_nodes = neo4j.find_nodes_by_label(neo4j_driver, "Week")
        if not week_nodes:
            logging.error("No week nodes found in the database.")
            return
    if node_types_options['CalendarDayNode']:
        logging.debug(f"Fetching day nodes...")
        day_nodes = neo4j.find_nodes_by_label(neo4j_driver, "Day")
        if not day_nodes:
            logging.error("No day nodes found in the database.")
            return
    
    # Create relationships based on the options
    if has_relationships:
        logging.debug(f"Creating 'HAS' relationships for calendar...")

        has_relationships_to_be_merged = []

        if node_types_options.get('CalendarYearNode', False):

            logging.debug(f"Number of calendar year nodes: {len(calendar_year_nodes)}")

            for calendar_year_node in calendar_year_nodes:

                if calendar_year_node is None or 'properties' not in calendar_year_node or calendar_year_node['properties'] is None:
                    logging.error(f"Calendar year node is None or missing 'properties': {calendar_year_node}")
                    continue
                
                calendar_year_start_date = calendar_year_node['properties']['start_date']
                calendar_year_end_date = calendar_year_node['properties']['end_date']
                
                if node_types_options.get('CalendarMonthNode', False):
                        
                    for month_node in month_nodes:

                        month_start_date = month_node['properties']['start_date']
                        month_end_date = month_node['properties']['end_date']

                        if date_in_range(month_start_date, calendar_year_start_date, calendar_year_end_date):

                            relationship = CalendarYearHasMonth(source=calendar_year_node, target=month_node)
                            has_relationships_to_be_merged.append(relationship)

                if node_types_options.get('CalendarWeekNode', False):

                    for week_node in week_nodes:

                        week_start_date = week_node['properties']['start_date']
                        week_end_date = week_node['properties']['end_date']

                        if date_in_range(week_start_date, calendar_year_start_date, calendar_year_end_date):

                            relationship = CalendarYearHasWeek(source=calendar_year_node, target=week_node)
                            has_relationships_to_be_merged.append(relationship)

                if node_types_options.get('CalendarDayNode', False):

                    for day_node in day_nodes:

                        date = day_node['properties']['date']

                        if date_in_range(date, calendar_year_start_date, calendar_year_end_date):

                            relationship = CalendarYearHasDay(source=calendar_year_node, target=day_node)
                            has_relationships_to_be_merged.append(relationship)

        if node_types_options.get('CalendarMonthNode', False):

            logging.debug(f"Number of month nodes: {len(month_nodes)}")

            for month_node in month_nodes:

                month_start_date = month_node['properties']['start_date']
                month_end_date = month_node['properties']['end_date']

                if node_types_options.get('WeekNode', False):

                    for week_node in week_nodes:

                        week_start_date = week_node['properties']['start_date']
                        week_end_date = week_node['properties']['end_date']

                        if date_in_range(week_start_date, month_start_date, month_end_date) or date_in_range(week_end_date, month_start_date, month_end_date):

                            relationship = MonthHasWeek(source=month_node, target=week_node)
                            has_relationships_to_be_merged.append(relationship)

                if node_types_options.get('CalendarDayNode', False):

                    for day_node in day_nodes:

                        date = day_node['properties']['date']

                        if date_in_range(date, month_start_date, month_end_date):

                            relationship = MonthHasDay(source=month_node, target=day_node)
                            has_relationships_to_be_merged.append(relationship)

        if node_types_options.get('CalendarWeekNode', False):

            logging.debug(f"Number of week nodes: {len(week_nodes)}")

            for week_node in week_nodes:

                week_start_date = week_node.get('start_date')
                week_end_date = week_node.get('end_date')
                
                if node_types_options.get('CalendarDayNode', False):

                    for day_node in day_nodes:

                        date = day_node.get('date')

                        if date_in_range(date, week_start_date, week_end_date):
                            
                            day_node_model = CalendarDayNode(**day_node)
                            week_node_model = CalendarWeekNode(**week_node)

                            relationship = WeekHasDay(source=week_node_model, target=day_node_model)
                            has_relationships_to_be_merged.append(relationship)

        if node_types_options.get('CalendarDayNode', False):

            logging.debug(f"Not implemented here.")


    if part_of_relationships:

        logging.debug(f"Creating 'PART_OF' relationships for calendar...")

        part_of_relationships_to_be_merged = []

        if node_types_options.get('CalendarYearNode', False):

            logging.error(f"Not implemented here.")

        if node_types_options.get('CalendarWeekNode', False):

            for week_node in week_nodes:

                if node_types_options.get('CalendarYearNode', False):

                    for calendar_year_node in calendar_year_nodes:
                        
                        relationship = WeekBelongsToCalendarYear(source=week_node, target=calendar_year_node)
                        part_of_relationships_to_be_merged.append(relationship)

        if node_types_options.get('CalendarDayNode', False):

            for day_node in day_nodes:

                if node_types_options.get('CalendarYearNode', False):

                    for calendar_year_node in calendar_year_nodes:
                        
                        relationship = DayBelongsToCalendarYear(source=day_node, target=calendar_year_node)
                        part_of_relationships_to_be_merged.append(relationship)

                if node_types_options.get('CalendarWeekNode', False):

                    for week_node in week_nodes:
                        
                        relationship = DayBelongsToWeek(source=day_node, target=week_node)
                        part_of_relationships_to_be_merged.append(relationship)

    logging.debug(f"'HAS' relationships to be merged: {has_relationships_to_be_merged}")
    logging.debug(f"'PART_OF' relationships to be merged: {part_of_relationships_to_be_merged}")

    if write:
        logging.info("Writing calendar relationships to the database...")
        for relationship in has_relationships_to_be_merged:
            neon.create_or_merge_neontology_relationship(relationship, operation='merge')
        for relationship in part_of_relationships_to_be_merged:
            neon.create_or_merge_neontology_relationship(relationship, operation='merge')

    return has_relationships_to_be_merged, part_of_relationships_to_be_merged
```

```python
# Create timetable relationships

def create_timetable_relationships(neo4j_driver, node_types_options, has_relationships, part_of_relationships, write=False):
    """
    Create relationships between nodes based on provided options.

    Args:
        neo4j_driver: The Neo4j driver for database connection.
        has_relationships (bool): Whether to create 'Has' relationships.
        part_of__relationships (bool): Whether to create 'BelongsTo' relationships.
        node_types_options (dict): A dictionary specifying which node types to apply the relationships to.
        write (bool): Whether to write changes to the database.
    """
    logging.info(f"Creating timetable relationships...")
    # Fetch nodes from the database based on the node types specified in the options
    if node_types_options['AcademicYearNode']:
        logging.debug(f"Fetching academic year nodes...")
        academic_year_nodes = neo4j.find_nodes_by_label(neo4j_driver, "AcademicYear")
        if not academic_year_nodes:
            logging.error("No academic year nodes found in the database.")
            return
    if node_types_options['TimetableNode']:
        logging.debug(f"Fetching timetable nodes...")
        timetable_nodes = neo4j.find_nodes_by_label(neo4j_driver, "Timetable")
        if not timetable_nodes:
            logging.error("No timetable nodes found in the database.")
            return
    if node_types_options['TermNode']:
        logging.debug(f"Fetching term nodes...")
        term_nodes = neo4j.find_nodes_by_label(neo4j_driver, "Term")
        if not term_nodes:
            logging.error("No term nodes found in the database.")
            return
    if node_types_options['AcademicDayNode']:
        logging.debug(f"Fetching academic day nodes...")
        academic_day_nodes = neo4j.find_nodes_by_label(neo4j_driver, "AcademicDay")
        if not academic_day_nodes:
            logging.error("No academic day nodes found in the database.")
            return
    if node_types_options['PeriodNode']:
        logging.debug(f"Fetching period nodes...")
        period_nodes = neo4j.find_nodes_by_label(neo4j_driver, "Period")
        if not period_nodes:
            logging.error("No period nodes found in the database.")
            return

    # Create relationships based on the options

    if has_relationships:
        logging.debug(f"Creating 'HAS' relationships for timetable...")

        has_relationships_to_be_merged = []

        if node_types_options.get('AcademicYearNode', False):

            logging.debug(f"Number of academic year nodes: {len(academic_year_nodes)}")

            for academic_year_node in academic_year_nodes:

                if academic_year_node is None or 'properties' not in academic_year_node or academic_year_node['properties'] is None:
                    logging.error(f"Academic year node is None or missing 'properties': {academic_year_node}")
                    continue
                
                academic_year_start_date = academic_year_node['properties']['academic_year_start_date']
                academic_year_end_date = academic_year_node['properties']['academic_year_end_date']
                
                if node_types_options.get('TermNode', False):
                        
                    for term_node in term_nodes:

                        term_start_date = term_node['properties']['term_start_date']
                        term_end_date = term_node['properties']['term_end_date']

                        if date_in_range(term_start_date, academic_year_start_date, academic_year_end_date):

                            relationship = AcademicYearHasTerm(source=academic_year_node, target=term_node)
                            has_relationships_to_be_merged.append(relationship)

                if node_types_options.get('AcademicDayNode', False):

                    for academic_day_node in academic_day_nodes:

                        academic_day_date = academic_day_node['properties']['date']

                        if date_in_range(academic_day_date, academic_year_start_date, academic_year_end_date):

                            relationship = AcademicYearHasAcademicDay(source=academic_year_node, target=academic_day_node)
                            has_relationships_to_be_merged.append(relationship)

        if node_types_options.get('TermNode', False):

            logging.debug(f"Number of term nodes: {len(term_nodes)}")

            for term_node in term_nodes:
                
                term_start_date = term_node.get('term_start_date')
                term_end_date = term_node.get('term_end_date')

                if node_types_options.get('AcademicDayNode', False):

                    for academic_day_node in academic_day_nodes:

                        academic_day_date = academic_day_node.get('date')

                        if date_in_range(academic_day_date, term_start_date, term_end_date):

                            relationship = TermHasAcademicDay(source=term_node, target=academic_day_node)
                            has_relationships_to_be_merged.append(relationship)

                if node_types_options.get('PeriodNode', False):

                    for period_node in period_nodes:

                        period_start_time = period_node.get('period_start_datetime')

                        if datetime_in_daterange(period_start_time, term_start_date, term_end_date):
                            
                            period_node_model = PeriodNode(**period_node)
                            term_node_model = TermNode(**term_node)

                            relationship = TermHasPeriod(source=term_node_model, target=period_node_model)
                            has_relationships_to_be_merged.append(relationship)

        if node_types_options.get('AcademicDayNode', False):

            logging.debug(f"Number of academic day nodes: {len(academic_day_nodes)}")

            for academic_day_node in academic_day_nodes:

                academic_day_date = academic_day_node.get('date')
                
                if node_types_options.get('PeriodNode', False):

                    for period_node in period_nodes:

                        period_start_time = period_node.get('period_start_datetime')

                        if datetime_in_daterange(period_start_time, academic_day_date, academic_day_date):
                            
                            period_node_model = PeriodNode(**period_node)
                            academic_day_node_model = AcademicDayNode(**academic_day_node)

                            relationship = AcademicDayHasPeriod(source=academic_day_node_model, target=period_node_model)
                            has_relationships_to_be_merged.append(relationship)

    if part_of_relationships:

        logging.debug(f"Creating 'PART_OF' relationships for timetable...")

        part_of_relationships_to_be_merged = []

        if node_types_options.get('AcademicYearNode', False):

            logging.error(f"Not implemented here.")

        if node_types_options.get('TermNode', False):

            for term_node in term_nodes:

                if node_types_options.get('AcademicYearNode', False):

                    for academic_year_node in academic_year_nodes:
                        
                        relationship = TermBelongsToAcademicYear(source=term_node, target=academic_year_node)
                        part_of_relationships_to_be_merged.append(relationship)

        if node_types_options.get('AcademicDayNode', False):

            for academic_day_node in academic_day_nodes:

                if node_types_options.get('AcademicYearNode', False):

                    for academic_year_node in academic_year_nodes:
                        
                        relationship = AcademicDayBelongsToAcademicYear(source=academic_day_node, target=academic_year_node)
                        part_of_relationships_to_be_merged.append(relationship)

                if node_types_options.get('TermNode', False):

                    for term_node in term_nodes:
                        
                        relationship = AcademicDayBelongsToTerm(source=academic_day_node, target=term_node)
                        part_of_relationships_to_be_merged.append(relationship)

        if node_types_options.get('PeriodNode', False):

            for period_node in period_nodes:

                if node_types_options.get('AcademicYearNode', False):

                    for academic_year_node in academic_year_nodes:
                        
                        relationship = PeriodBelongsToAcademicYear(source=period_node, target=academic_year_node)
                        part_of_relationships_to_be_merged.append(relationship)

                if node_types_options.get('TermNode', False):

                    for term_node in term_nodes:
                        
                        relationship = PeriodBelongsToTerm(source=period_node, target=term_node)
                        part_of_relationships_to_be_merged.append(relationship)

                if node_types_options.get('AcademicDayNode', False):

                    for academic_day_node in academic_day_nodes:
                        
                        relationship = PeriodBelongsToAcademicDay(source=period_node, target=academic_day_node)
                        part_of_relationships_to_be_merged.append(relationship)

    logging.debug(f"'HAS' relationships to be merged: {has_relationships_to_be_merged}")
    logging.debug(f"'PART_OF' relationships to be merged: {part_of_relationships_to_be_merged}")

    # Finally, write the relationships to the database if `write` is True
    if write:
        logging.info("Writing timetable relationships to the database...")
        for relationship in has_relationships_to_be_merged:
            neon.create_or_merge_neontology_relationship(relationship, operation='merge')
        for relationship in part_of_relationships_to_be_merged:
            neon.create_or_merge_neontology_relationship(relationship, operation='merge')

    return has_relationships_to_be_merged, part_of_relationships_to_be_merged
```

```python
# Timeflow helper functions

# Function to trim the 'Node' suffix from a node name
def trim_node_from_end(name):
    return name[:-4]
```

```python
# Timeflow functions

# Function to create timeflow relationships between nodes
def create_timeflow_relationships(neo4j_driver, has_next, has_previous, node_types_options, write=False):
    """ 
    Create timeflow relationships between nodes based on provided options.
    Args:
        neo4j_driver: The Neo4j driver for database connection.
        node_types_options (dict): A dictionary specifying which node types to apply the relationships to.
        write (bool): Whether to write changes to the database.
    """
    # Fetch nodes from the database based on the node types specified in the options
    nodes = {}
    for node_type in node_types_options:
        logging.debug(f"Node type: {node_type}")
        if node_types_options[node_type]:
            logging.debug(f"Node type {node_type} is in the options.")
            node_label = trim_node_from_end(node_type)
            logging.debug(f"Fetching {node_type} nodes...")
            nodes[node_type] = neo4j.find_nodes_by_label(neo4j_driver, node_label)
            if not nodes[node_type]:
                logging.error(f"No {node_type} found in the database.")
                return

    # Sort nodes by date
    for node_type in nodes:
        logging.debug(f"Node type: {node_type}")
        if node_type in ['TermNode']:
            logging.debug(f"Sorting {node_type} nodes by term_start_date...")
            nodes[node_type] = sorted(nodes[node_type], key=lambda x: x.get('term_start_date'))
        elif node_type in ['TermBreakNode']:
            nodes[node_type] = sorted(nodes[node_type], key=lambda x: x.get('term_break_start_date'))
        elif node_type in ['WeekNode']:
            nodes[node_type] = sorted(nodes[node_type], key=lambda x: x.get('week_start_date'))
        elif node_type in ['DateNode']:
            nodes[node_type] = sorted(nodes[node_type], key=lambda x: x.get('date'))
        elif node_type in ['PeriodNode']:
            nodes[node_type] = sorted(nodes[node_type], key=lambda x: x.get('period_start_datetime'))

    # Create timeflow relationships based on the options
    logging.info(f"Creating timeflow relationships...")
    relationships_to_be_merged = []
    for node_type in node_types_options:
        logging.debug(f"Node type: {node_type}")
        if node_types_options[node_type]:
            logging.debug(f"Node type {node_type} is in the options.")
            for i in range(len(nodes[node_type])-1):
                source_node_properties = nodes[node_type][i]
                target_node_properties = nodes[node_type][i+1]

                source_model = globals()[node_type](**source_node_properties)
                target_model = globals()[node_type](**target_node_properties)

                if has_next:
                    node_type_trimmed = trim_node_from_end(node_type) # For example, 'TermNode'
                    relationship_class = globals()[f"{node_type_trimmed}HasNext{node_type_trimmed}"]
                    relationship = relationship_class(source=source_model, target=target_model)
                    relationships_to_be_merged.append(relationship)

                if has_previous:
                    node_type_trimmed = trim_node_from_end(node_type) # For example, 'TermNode'
                    relationship_class = globals()[f"{node_type_trimmed}HasPrevious{node_type_trimmed}"]
                    relationship = relationship_class(source=target_model, target=source_model)
                    relationships_to_be_merged.append(relationship)

    # Write changes to the database
    if write:
        for relationship in relationships_to_be_merged:
            neon.create_or_merge_neontology_relationship(relationship, operation='merge')

    return relationships_to_be_merged
```

```python
## Curriculum layer relationship functions

# TopicNode and TopicLessonNode
def create_topic_has_lesson_relationships(neo4j_driver, topic_lesson_excel_node_list, options, write=False):
    logging.info(f"Creating topic_has_lesson relationships from {len(topic_lesson_excel_node_list)} topic lesson data...")

    relationship_list = []

    for topic_lesson_node in topic_lesson_excel_node_list:
        
        topic_lesson_node_id = topic_lesson_node['node'].id

        lesson_nodes = neo4j.find_nodes_by_label_and_properties(neo4j_driver, 'TopicLesson', {'id': topic_lesson_node_id})

        if not lesson_nodes:  # Check if the list is empty (no nodes found), if so, log an error and continue
            logging.error(f"No lesson node found for ID {topic_lesson_node_id}")
            continue

        topic_node_id = topic_lesson_node['helper']['topic_lesson_topic_id']
        
        topic_nodes = neo4j.find_nodes_by_label_and_properties(neo4j_driver, 'Topic', {'id': topic_node_id})
        
        if not topic_nodes:  # Check if the list is empty, if so, log an error and continue
            logging.error(f"No topic node found for ID {topic_node_id}")
            continue
        
        # Assuming only one node is expected for each query (or no nodes at all, which we need to check for)
        lesson_node_properties = dict(lesson_nodes[0])  # Extract properties from the Neo4j Node
        topic_node_properties = dict(topic_nodes[0])    # Extract properties from the Neo4j Node
        

        # Create instances of TopicNode and TopicLessonNode using the extracted properties
        neontology_lesson_node = TopicLessonNode(**lesson_node_properties)
        neontology_topic_node = TopicNode(**topic_node_properties)
        

        # Create the relationship
        topic_has_lesson_relationship = TopicIncludesTopicLesson(source=neontology_topic_node, target=neontology_lesson_node)

        relationship_list.append(topic_has_lesson_relationship)

    # If write=True merge or create relationship in the database
    if write:
        logging.info(f"Writing {len(relationship_list)} topic_has_lesson relationships to the database...")
        for relationship in relationship_list:
            neon.create_or_merge_neontology_relationship(relationship, operation='merge')

    return relationship_list

# TopicNode and LearningStatementNode and LessonNode and LearningStatementNode
def create_learning_statement_relationships(neo4j_driver, learning_statement_excel_node_list, options, write=False):
    logging.info(f"Creating learning statement relationships from {len(learning_statement_excel_node_list)} learning statement data...")
    
    statement_in_lesson_relationship_list = []
    lesson_has_statement_relationship_list = []
    learning_statement_in_topic_relationship_list = []
    topic_has_learning_statement_relationship_list = []
    learning_statement_relationships_list = []
    
    for learning_statement_node in learning_statement_excel_node_list:
        logging.debug(f"Processing learning statement node: {learning_statement_node}")

        learning_statement_node_id = learning_statement_node['node'].id
        
        learning_statement_nodes = neo4j.find_nodes_by_label_and_properties(neo4j_driver, 'LearningStatement', {'id': learning_statement_node_id})
        
        if not learning_statement_nodes:  # Check if the list is empty
            logging.error(f"No learning statement node found for ID {learning_statement_node_id}")
            continue
        
        learning_statement_node_properties = dict(learning_statement_nodes[0])  # Extract properties from the Neo4j Node
        
        topic_node_id = learning_statement_node['helper']['lesson_learning_statement_topic_id']
        
        topic_nodes = neo4j.find_nodes_by_label_and_properties(neo4j_driver, 'Topic', {'id': topic_node_id})

        if not topic_nodes:  # Check if the list is empty
            logging.error(f"No topic node found for ID {topic_node_id}")
            continue
        
        topic_node_properties = dict(topic_nodes[0])    # Extract properties from the Neo4j Node

        topic_lesson_node_id = learning_statement_node['helper']['lesson_learning_statement_lesson_id']
        
        topic_lesson_nodes = neo4j.find_nodes_by_label_and_properties(neo4j_driver, 'TopicLesson', {'id': topic_lesson_node_id})
        
        if not topic_lesson_nodes:
            logging.error(f"No topic lesson node found for ID {topic_lesson_node_id}")
            continue
        
        lesson_node_properties = dict(topic_lesson_nodes[0])  # Extract properties from the Neo4j Node
        
        # Create instances of TopicNode and TopicLessonNode using the extracted properties
        neontology_lesson_node = TopicLessonNode(**lesson_node_properties)
        neontology_topic_node = TopicNode(**topic_node_properties)
        neontology_learning_statement_node = LearningStatementNode(**learning_statement_node_properties)
        
        if options['LessonStatements']:
            # Create the relationship between learning statement and lesson
            statement_in_lesson_relationship = LearningStatementPartOfTopicLesson(source=neontology_learning_statement_node, target=neontology_lesson_node)
            lesson_has_statement_relationship = TopicLessonIncludesLearningStatement(source=neontology_lesson_node, target=neontology_learning_statement_node)
            statement_in_lesson_relationship_list.append(statement_in_lesson_relationship)
            lesson_has_statement_relationship_list.append(lesson_has_statement_relationship)

        if options['TopicStatements']:
            topic_has_learning_statement_relationship = TopicIncludesLearningStatement(source=neontology_topic_node, target=neontology_learning_statement_node)
            learning_statement_in_topic_relationship = LearningStatementPartOfTopic(source=neontology_learning_statement_node, target=neontology_topic_node)
            topic_has_learning_statement_relationship_list.append(topic_has_learning_statement_relationship)
            learning_statement_in_topic_relationship_list.append(learning_statement_in_topic_relationship)

        learning_statement_relationships_list.append(statement_in_lesson_relationship_list)

    # If Write=True merge or create relationships in the database
    if write:
        if options['LessonStatements']:
            logging.info(f"Writing {len(statement_in_lesson_relationship_list)} statement in lesson relationships to the database...")

            for relationship in statement_in_lesson_relationship_list:
                neon.create_or_merge_neontology_relationship(relationship, operation='merge')

            logging.info(f"Writing {len(lesson_has_statement_relationship_list)} lesson has statement relationships to the database...")

            for relationship in lesson_has_statement_relationship_list:
                neon.create_or_merge_neontology_relationship(relationship, operation='merge')

        if options['TopicStatements']:
            logging.info(f"Writing {len(learning_statement_in_topic_relationship_list)} learning statement in topic relationships to the database...")

            for relationship in learning_statement_in_topic_relationship_list:
                neon.create_or_merge_neontology_relationship(relationship, operation='merge')

            logging.info(f"Writing {len(topic_has_learning_statement_relationship_list)} topic has learning statement relationships to the database...")

            for relationship in topic_has_learning_statement_relationship_list:
                neon.create_or_merge_neontology_relationship(relationship, operation='merge')

    return learning_statement_relationships_list
```

```python
# Create relationships between ALL SchoolNodes and DepartmentNodes
def create_school_has_department_relationships(neo4j_driver, schools, departments, options, write=False):
    logging.info(f"Creating school has department relationships from {len(schools)} schools and {len(departments)} departments...")
    
    school_has_department_relationship_list = []

    for school_node in schools:
        
        school_node_id = school_node.school_id
        
        school_nodes = neo4j.find_nodes_by_label_and_properties(neo4j_driver, 'School', {'school_id': school_node_id})
        
        if not school_nodes:  # Check if the list is empty
            logging.error(f"No school node found for ID {school_node_id}")
            continue

        school_node_properties = dict(school_nodes[0])  # Extract properties from the Neo4j Node

        for department_node in departments:

            department_node_id = department_node.department_id
            
            department_nodes = neo4j.find_nodes_by_label_and_properties(neo4j_driver, 'Department', {'department_id': department_node_id})
            
            if not department_nodes:  # Check if the list is empty
                logging.error(f"No department node found for ID {department_node_id}")
                continue

            department_node_properties = dict(department_nodes[0])  # Extract properties from the Neo4j Node

            neontology_school_node = SchoolNode(**school_node_properties)
            neontology_department_node = DepartmentNode(**department_node_properties)

            school_has_department_relationship = SchoolHasDepartment(source=neontology_school_node, target=neontology_department_node)

            school_has_department_relationship_list.append(school_has_department_relationship)

    # Merge or create relationship in the database
    if write:
        logging.info(f"Writing {len(school_has_department_relationship_list)} school has department relationships to the database...")

        for relationship in school_has_department_relationship_list:
            neon.create_or_merge_neontology_relationship(relationship, operation='merge')

    return school_has_department_relationship

school_has_department_relationship = create_school_has_department_relationships(neo4j_driver, schools, departments, options, write=True)
```

```python
# Function to create a date object from a string

def create_date_from_string(date_string):
    date_time = datetime.strptime(date_string, '%Y-%m-%d')
    return date_time.date()

def create_learning_statement_relationships_for_planned_lesson(planned_lesson_node, topic_lesson_node, learning_statement_excel_node_list, write=False):
    logging.info(f"Creating custom relationships between planned lesson and learning statement from {len(learning_statement_excel_node_list)} learning statement data...")

    planned_lesson_relationships_list = []
    learning_statement_relationships_list = []
    relationships_list = []

    planned_lesson_node_properties = dict(planned_lesson_node)
    logging.debug(f"Planned lesson node properties: {planned_lesson_node_properties}")

    # Iterate over each LearningStatementNode data from the excel node list
    for learning_statement_data in learning_statement_excel_node_list:
        
        # If the learning staement is for the topic lesson, create a relationship
        if learning_statement_data['helper']['lesson_learning_statement_lesson_id'] == topic_lesson_node.id:
            logging.info(f"Creating relationship between planned lesson and learning statement for learning statement {learning_statement_data['node'].id} and planned lesson {planned_lesson_node}")
            learning_statement_node = learning_statement_data['node']

            # Create a relationship since it's a match
            logging.debug(f"Creating relationship between planned lesson and learning statement for learning statement {learning_statement_node} and planned lesson {planned_lesson_node}")
            learning_statement_is_for_planned_lesson_relationship = LearningStatementIsForPlannedLesson(source=learning_statement_node, target=planned_lesson_node_properties)
            logging.debug(f"Creating relationship between planned lesson and learning statement for planned lesson {planned_lesson_node} and learning statement {learning_statement_node}")
            planned_lesson_is_about_learning_statement_relationship = PlannedLessonIsAboutLearningStatement(source=planned_lesson_node_properties, target=learning_statement_node)

            planned_lesson_relationships_list.append(learning_statement_is_for_planned_lesson_relationship)
            learning_statement_relationships_list.append(planned_lesson_is_about_learning_statement_relationship)

            relationships_list.append(learning_statement_is_for_planned_lesson_relationship)
            relationships_list.append(planned_lesson_is_about_learning_statement_relationship)

    if write:
        logging.info(f"Writing {len(learning_statement_is_for_planned_lesson_relationship)} learning statement has planned lesson relationships to the database...")
        neon.create_or_merge_neontology_relationship(learning_statement_is_for_planned_lesson_relationship, operation='merge')
        logging.info(f"Writing {len(planned_lesson_is_about_learning_statement_relationship)} planned lesson is about learning statement relationships to the database...")
        neon.create_or_merge_neontology_relationship(planned_lesson_is_about_learning_statement_relationship, operation='merge')

    return relationships_list

def create_custom_relationships_between_planned_and_topic_lessons(neo4j_driver, planned_lesson_options, topic_lesson_excel_node_list, write=False):
    logging.info(f"Creating custom relationships between planned and topic lessons from {len(topic_lesson_excel_node_list)} topic lesson data...")

    planned_lesson_relationships_list = []
    topic_lesson_relationships_list = []
    learning_statement_relationships_list = []

    relationships_list = []

    # Assuming planned_lesson_options is a list of dictionaries
    for option in [planned_lesson_options]:  # Wrapping the single dictionary in a list for iteration
        staff_code = option['staff_code']
        date = option['date']
        period_name = option['period_name']
        topic_lesson_id = option['id']
        
        # Find PlannedLesson nodes matching the criteria
        planned_lesson_nodes = neo4j.find_nodes_by_label_and_properties(neo4j_driver, 'Lesson', {'teacher_id': staff_code, 'date': date, 'timetable_period_name': period_name})

        if not planned_lesson_nodes:
            logging.error(f"No planned lesson node found for teacher code {staff_code}, date {date} and period name {period_name}")
            continue
        
        for planned_lesson_node in planned_lesson_nodes:
            planned_lesson_properties = dict(planned_lesson_node)

            # Iterate over each TopicLessonNode data from the excel node list
            for topic_lesson_data in topic_lesson_excel_node_list:
                topic_lesson_node = topic_lesson_data['node']

                # Check if the TopicLessonNode ID matches the topic_lesson_id from the options
                if topic_lesson_node.id == topic_lesson_id:
                    # Create a relationship since it's a match
                    logging.debug(f"Creating relationship between planned lesson and topic lesson for topic lesson {topic_lesson_node} and planned lesson {planned_lesson_node}")
                    topic_lesson_is_for_planned_lesson_relationship = TopicLessonIsForPlannedLesson(source=topic_lesson_node, target=planned_lesson_properties)
                    logging.debug(f"Creating relationship between planned lesson and topic lesson for planned lesson {planned_lesson_node} and topic lesson {topic_lesson_node}")
                    planned_lesson_is_about_topic_lesson_relationship = PlannedLessonIsAboutTopicLesson(source=planned_lesson_properties, target=topic_lesson_node)

                    # Create relationships between planned lesson and learning statement
                    logging.info(f"Creating learning statement relationships for planned lesson {planned_lesson_node} and topic lesson {topic_lesson_node}")
                    learning_staement_relationships = create_learning_statement_relationships_for_planned_lesson(planned_lesson_node, topic_lesson_node, learning_statement_excel_node_list, write=False)

                    planned_lesson_relationships_list.append(topic_lesson_is_for_planned_lesson_relationship)
                    topic_lesson_relationships_list.append(planned_lesson_is_about_topic_lesson_relationship)
                    learning_statement_relationships_list.append(learning_staement_relationships)

                    relationships_list.append(topic_lesson_is_for_planned_lesson_relationship)
                    relationships_list.append(planned_lesson_is_about_topic_lesson_relationship)

                    if write:
                        logging.info(f"Writing topic lesson has planned lesson relationships {topic_lesson_is_for_planned_lesson_relationship} to the database...")
                        neon.create_or_merge_neontology_relationship(topic_lesson_is_for_planned_lesson_relationship, operation='merge')
                        logging.info(f"Writing planned lesson is about topic lesson relationships for {planned_lesson_is_about_topic_lesson_relationship} to the database...")
                        neon.create_or_merge_neontology_relationship(planned_lesson_is_about_topic_lesson_relationship, operation='merge')
                        logging.info(f"Writing learning statement relationships for {learning_staement_relationships} to the database...")
                        for relationship in learning_staement_relationships:
                            neon.create_or_merge_neontology_relationship(relationship, operation='merge')

    return relationships_list
```
